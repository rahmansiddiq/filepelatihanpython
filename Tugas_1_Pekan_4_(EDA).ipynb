```python
# code here
df=pd.read_csv('austin_weather.csv',na_values=['na','-'])
df.astype({'DewPointAvgF': 'float','HumidityAvgPercent':'float','WindAvgMPH':'float'}).dtypes
```

TANPA HANDLING MISSING VALUES

```python
import seaborn as sns
sns.set({'figure.figsize':(12,9)})
sns.scatterplot(x='HumidityAvgPercent',y='DewPointAvgF',data=df,hue='TempAvgF',size='WindAvgMPH',sizes=(150,150),linewidth=0,palette='coolwarm').set(title='Austin Weather');
```

MENGGUNAKAN HANDLING MISSING VALUES

```python
df1=df.fillna(method='ffill')
df1.info()
```

```python
sns.set({'figure.figsize':(12,9)})
sns.scatterplot(x='HumidityAvgPercent',y='DewPointAvgF',data=df1,hue='TempAvgF',size='WindAvgMPH',sizes=(150,150),linewidth=0,palette='coolwarm').set(title='Austin Weather');
```
