<h2>Soal 2: Data Restructuring</h2>

Jelaskan itu pivoting dan melting data

Pivoting adalah suatu pengubahan bentuk data dengan memutar data yang terletak di baris menjadi di column, dan juga dapat melakukan perubahan pada bentuk data dari yang tadinya panjang menjadi lebar

---
Melting adalah ‘unpivoting’, melting mengubah suatu data dengan memutar dari yang tadinya berada di posisi column menjadi di posisi row. Melting mengubah data yang tadinya lebar menjadi panjang.
Melting berguna ketika ada suatu kasus dimana terdapat satu atau lebih column yang merupakan identifier variable, dan columnn lain adalah measure variabel.

<h2>Soal 3: Reshaping Data 1</h2>

```python
import pandas as pd
import random

data = {
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)] 
}

df = pd.DataFrame(data)
df
```

Perhatikan data di atas, kemudian ubahlah bentuknya agar sesuai yang di harapkan.

```python
#code here
pivot_df=pd.pivot(df,index='Tipe Ujian',columns='Mata Pelajaran',values=['Rata2 Nilai Ujian','Rata2 Waktu Belajar (menit)'])
pivot_df
```

<h2>Soal 4: Reshaping Data 2</h2>

```python
import pandas as pd

data = {
    'Hari' : ['Sabtu', 'Minggu'],
    'Kebun Binatang' : [271, 399],
    'Dufan' : [501, 700],
    'Disney' : [1000, 1001],
    'Bali' : [900, 803]
}
df = pd.DataFrame(data).set_index('Hari')
df
```

diatas adalah data tentang jumlah pengunjung di berbagai tempat hiburan, ubahlah struktur data tersebut sehingga sesuai dengan hasil yang diharapkan.

```python
#code here
df_reset=df.reset_index()
df_reset
```

```python
pd.melt(df_reset,id_vars=['Hari'],value_vars=['Kebun Binatang','Dufan','Disney','Bali'],
       var_name='Tempat Hiburan',value_name='Pengunjung')
```

<h2>Soal 5: Data Enrichment</h2>

Lakukan binning data menggunakan qcut sehingga sesuai hasil yang diharapkan

```python
#code here
data={
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)] 
}
df_data = pd.DataFrame(data)
df_data
```

```python
df_data['Keterangan Kelulusan']=df_data['Rata2 Nilai Ujian']
df_data
```

```python
pd.qcut(df_data['Keterangan Kelulusan'],q=3)
```

```python
labels=['tidak lulus','remedial','lulus']
df_data['Keterangan Kelulusan']=pd.qcut(df_data['Keterangan Kelulusan'],q=3,labels=labels)
df_data
```
