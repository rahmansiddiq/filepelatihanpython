```python
#data restructing
#pivoting data=suatu pengubahan bentuk data dengan memutar data yang terletak di baris menjadi di column
data={'foo':['one']*3+['two']*3,
     'bar':['A','B','C','A','B','C'],
     'baz':[1,2,3,4,5,6],
     'zoo':['x','y','z','q','w','t']}
df_data=pd.DataFrame(data)
df_data
```

```python
#using pivot
pivot_data=pd.pivot(df_data,index='foo',columns='bar',values='baz')
pivot_data
```

```python
#using pivot table
pivot_table_data=pd.pivot_table(df_data,index='foo',columns='bar',values='baz',aggfunc='sum')
pivot_table_data
```

```python
#Perbedaan antara pivot dan pivot_table terletak pada penggunaan fungsi agregasi.
#Dalam pivot, tidak bisa dilakukan agregasi
#sedangkan pada pivot_table dapat diaplikasikan fungsi agregasi.
```

```python
#melting data=unpivoting, memutar data dari kolom menjadi row
#melting mengubah data yang tadinya lebar menjadi panjang
#Melting berguna ketika ada suatu kasus dimana terdapat satu atau lebih column yang merupakan identifier variable, dan columnn lain adalah measure variabel.
#Identifier variabel adalah suatu entitias yang melakukan identifikasi terhadap suatu variabel yang sedang di ukur.
```

```python
#membuat dataframe sederhan
df=pd.DataFrame({'Name':{0:'John',1:'Bob',2:'Shiela'},
                'Course':{0:'Masters',1:'Graduate',2:'Graduate'},
                'Age':{0:27,1:23,2:21}
                })
df
```

```python
#unpivot table
pd.melt(df,id_vars=['Name'],value_vars=['Course','Age'])
```
