```python
#MERGING DAN GROUPBY
#Data Merging=penggabungan data dengan berbagai macam logik
#beberapa method2 yg digunakan, join(), merge(),concat()
```

```python
#penggabungan baris dan kolom dengan method concat()
#concat= salah satu metode untuk melakukan penggabungan data berdasarkan sumbu tertentu.
import pandas as pd
df1=pd.DataFrame({'id_siswa':['A001','A002','A003','A004'],
                 'Kelas':['12A','12B','12C','12D'],
                 'Nilai':[100,98,96,95],
                 'Peringkat':[1,2,3,4]},
                 index=[0,1,2,3])
df2=pd.DataFrame({'id_siswa':['A005','A006','A007','A008'],
                 'Kelas':['12E','12F','12G','12H'],
                 'Nilai':[90,88,86,85],
                 'Peringkat':[5,6,7,8]},
                 index=[4,5,6,7])
df3=pd.DataFrame({'id_siswa':['A009','A010','A011','A012'],
                 'Kelas':['12I','12J','12K','12L'],
                 'Nilai':[80,75,73,70],
                 'Peringkat':[9,10,11,12]},
                 index=[8,9,10,11])
display(df1,df2,df3)
```

```python
#pertama kita buat terlebih dahulu list dari data yang akan kita gabung, kemudian masukan list tersebut sebagai argument kedalam method concat().
#list data yang akan digabung
list_dataframe=[df1,df2,df3]
#gabungkan dengan metode concat dari pandas
pd.concat(list_dataframe)
```

```python
#Apabila kita ingin menggabungkan data tersebut berdasarkan column, kita mesti memberikan satu argument lagi kedalam method concat yaitu axis, dengan nilai 1. seperti ini
pd.concat(list_dataframe,axis=1)
```

```python
#logika2 penggabungan dengan metode merge
#full outer join
#Full Outer Join menghasilkan gabungan index bari dari setiap data yang di gabung.
#Full Outer Join menggabungan pula column yang berbeda
#Memberikan nilai null untuk nilai column yang tidak di miliki oleh suatu index tertentu.
#misalkan
df1=pd.DataFrame({'id_siswa':['A001','A002','A003','A004'],
                 'Kelas':['12A','12B','12C','12D'],
                 'Tinggi Siswa (cm)':[155,160,170,165]},
                index=[0,1,2,3])
df2=pd.DataFrame({'id_siswa':['A001','A002','A005','A006'],
                 'Nilai':[100,98,96,95],
                 'Peringkat':[1,2,3,4]},
                index=[4,5,6,7])
print('data pertama')
print('=========')
print(df1)
print('data kedua')
print('=========')
print(df2)
```

```python
#akan digabungkan dengan metode merge
#column key dari data kita sebagai variabel identifier dari data kita, atau sebagai column key/referensi.
#Untuk melakukan penggabungan dengan menggunakan column key sebagai referensi penggabungan, maka kita akan memberikan argument on dengan nilai 'id_siswa'
#untuk melakukan penggabungan outer join kita akan memberikan argument how dengan nilai 'outer
#gabungkan data dengan logika outer join
pd.merge(df1,df2,on='id_siswa',how='outer')
```

```python
#gabungkan data dengan logika inner join
#Inner join ini akan menggabungkan data antara nilai key/referensi yang beririsan.
pd.merge(df1,df2,on='id_siswa',how='inner')
```

```python
#gabungkan data dengan left join
#Left Outer Join menghasilkan seluruh data dari data yang kiri ditambah data kanan yang memiliki nilai key yang sama dengan data dari kiri.
pd.merge(df1,df2,on='id_siswa',how='left')
```

```python
#gabungkan data dengan right join
#kebalikan dari left join
pd.merge(df1,df2,on='id_siswa',how='right')
```

```python
#Data Groupby
#grouping=suatu teknik memisahkan data berdasarkan kriteria tertentu
#ketika melakukan grouping ada beberapa proses yang terjadi secara berurutan, yaitu :
#Splitting : Memisahkan data kedalam suatu group berdasarkan kriteria tertentu.
#Applying : Melakukan suatu operasi terhadap sekumpulan data di group-group tersebut.
#Combining : Menggabungkan data menjadi suatu struktur baru
import numpy as np
import pandas as pd
df=pd.DataFrame([('bird','falconiformes',389.0),
                ('bird','psittaciformes',24.0),
                ('mammal','carnivora',80.2),
                ('mammal','primates',np.nan),
                ('mammal','carnivora',58)],
               index=['falcon','parrot','lion','monkey','leopard'],
               columns=('class','order','max_speed'))
df
```

```python
#misalkan group data tsb berdasarkan class
#proses splitting
df.groupby('class')
```

```python
#proses applying
df.groupby('class').mean()
```

```python
#proses combine
df.groupby(['class','order'])[['max_speed']].agg(['mean','sum'])
```
