<h2>Soal 1: Multiple Subplots</h2>

```python
#code here
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
fig,ax=plt.subplots(1,3,figsize=(20,8),sharex=True,sharey=True)
ax[0].plot(x,tan,label='tan x',color='blue')
ax[1].plot(x,sin,label='sin x',color='red')
ax[2].plot(x,cos,label='cos x',color='green')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig,ax=plt.subplots(figsize=(12,9))
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
ax.plot(x,y)
ax.set_title('Grafik Cos x')
ins1=ax.inset_axes([-0.3,-0.2,0.4,0.3])
ins1.set_title('Grafik Cos x^2')
ins1.plot(x,y,color='blue')
ins1.plot(x,y2,color='red')
ins2=ax.inset_axes([0.9,-0.2,0.4,0.3])
ins2.set_title('Grafik Cos x^3')
ins2.plot(x,y,color='blue')
ins2.plot(x,y3,color='green')
ins3=ax.inset_axes([0.9,0.9,0.4,0.3])
ins3.set_title('Grafik Cos x^4')
ins3.plot(x,y,color='blue')
ins3.plot(x,y4,color='black')
ins4=ax.inset_axes([-0.3,0.9,0.4,0.3])
ins4.set_title('Grafik Cos x^5')
ins4.plot(x,y,color='blue')
ins4.plot(x,y5,color='gold')
plt.show()
```
